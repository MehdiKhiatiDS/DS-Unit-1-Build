```
def number_to_direction(number):
    if number  in range (0,46):
        return 'NNE'
    elif number in range (46,91):
        return 'NE'
    elif number in range (91,136):
        return 'SW'
    elif number in range (136,181):
        return 'SSE'
    elif number in range (181,226):
        return 'SSW'
    elif number in range (226,271):
        return 'SW'
    elif number in range (271,316):
        return 'NW'
    elif number in range (316,361):
        return 'NNW'
```

```
upwelling['Wind Direction coordinates'] = upwelling['Wind Direction'].apply(number_to_direction)
upwelling['Wind Direction coordinates']
```
